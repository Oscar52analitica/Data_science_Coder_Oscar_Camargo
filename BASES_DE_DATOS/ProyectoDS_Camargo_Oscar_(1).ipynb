```python
# Cargar los datos
X = df_cars.drop(['price'],axis=1)
y = df_cars['price']
```

```python
# split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=16)

print('x_train size',X_train.shape)
print('y_train size',y_train.shape)
print('x_test size',X_test.shape)
print('y_test size',y_test.shape)
```

<a id="1"></a>
# <div style="padding:20px;color:white;margin:0;font-size:35px;font-family:Verdana;text-align:center;display:fill;border-radius:5px;background-color:#254E58;overflow:hidden"><font color=#4e8c4ee><b> 8. Hiperparametrosüìä</b></div>

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.linear_model import Ridge, Lasso
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor

# Ridge Regression
ridge_params = {'alpha': [0.01, 0.1, 1, 10, 100]}
ridge_grid = GridSearchCV(Ridge(), ridge_params, cv=5, scoring='r2')
ridge_grid.fit(X_train, y_train)
print("Mejor par√°metro para Ridge:", ridge_grid.best_params_)

# Lasso Regression
lasso_params = {'alpha': [0.01, 0.1, 1, 10, 100]}
lasso_grid = GridSearchCV(Lasso(), lasso_params, cv=5, scoring='r2')
lasso_grid.fit(X_train, y_train)
print("Mejor par√°metro para Lasso:", lasso_grid.best_params_)

# KNN Regressor
knn_params = {'n_neighbors': range(1, 30)}
knn_grid = GridSearchCV(KNeighborsRegressor(), knn_params, cv=5, scoring='r2')
knn_grid.fit(X_train, y_train)
print("Mejor par√°metro para KNN:", knn_grid.best_params_)

# XGBoost Regressor
xgb_params = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7]
}
xgb_random = RandomizedSearchCV(XGBRegressor(), xgb_params, cv=5, scoring='r2', n_iter=10, random_state=42)
xgb_random.fit(X_train, y_train)
print("Mejor par√°metro para XGBoost:", xgb_random.best_params_)
```
